### [주제]
지방소멸의 요인 중 하나로 '일자리'가 지목된다. 지역에 따른 일자리의 수 - 즉, 양적 차이는 분명해 보인다. 
#### 그렇다면 지역에 따른 일자리의 질적인 차이도 존재하는가?

이 자료 (https://sgis.kostat.go.kr/view/workRoad/main)는 워크넷, 인크루트, 사람인 등의 구인공고 사이트를 기반으로 지역별 구인 업체 수 및 인원 수를 보여준다. 이를 통해 지역에 따른 일자리 수의 차이를 확인할 수 있다.  

이는 지역별 총인구수(ref. https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1B040A3) 차이를 감안하더라도 유의미한 수준이다.

그렇다면 질적인 측면에서도 지역마다 차이가 존재할까?  
즉. "수도권일수록 소위 괜찮은 직업/직장이 많고 지방권일수록 괜찮은 직업/직장이 적을 것인가?" 에 대해 세 가지 가설을 통해 답을 구하고자 한다.

1. 구인 공고 데이터로 지역 예측이 가능할 것이다 (지도 학습).
지역에 따라 일자리의 질적인 수준이 명확하게 나뉠수록, 구인 공고 데이터를 모아서 지역을 예측할 때 정확도가 높은 모델을 만들기 쉬울 것이다.

2. 구인 공고 데이터 피쳐를 바탕으로 클러스터링 시에 크게 그룹 {우수한 일자리 / 중간 / 상대적으로 별로인 일자리} 로  나눌  수 있을 것이며, 우수한 일자리 그룹일수록 수도권에 해당하는 데이터포인트가 다수 분포할 것이다. (비지도 학습)

3. {수도권, 비수도권 광역시, 그 외 지방} 지역 중에서, <수도권> 지역이 '대기업 및 중견기업이 내건 구인공고 비중'이 가장 크고 <그 외 지방> 지역이 해당 공고 비중이 가정 작을 것이다.
