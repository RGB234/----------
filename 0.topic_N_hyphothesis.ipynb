### [주제] 지역에 따라서 (청년층의) 일자리에 대한 질적인 차이가 존재하는가?


전체 인구의 절반 가까이가 수도권에 모여산다는 것을 감안하면, 수도권 지역에서 구인구직이 전체 중 다수 이뤄진다는 것은 당연한 사실이다. 
또한 자료(https://sgis.kostat.go.kr/view/workRoad/main) 상으로도 해당 사실을 확인할 수 있다.  
따라서 지방소멸의 원인 중 하나로 '일자리 부족' 이 지목되려면, 다음 두 가설 중 하나 이상이 증명되어야 한다.  

1. 인구수 대비 구인 발생 건수를 비교하여, 해당 지표상 지방보다 수도권이 우수하다.  
2. 제안하는 연봉 및 근무조건이 전체적으로 지방보다 수도권이 우수하다.  

그리고 둘 중에서 1번 대신, 데이터 수집 및 분석 역량이 더 요구되는 2번에 대해 분석을 수행하기로 결정하였다.

구체적으로 지방 소멸의 원인 중 하나를 '구인구직 활동을 위해 지방을 이탈하는 청년층이 다수 존재하기 때문' 으로 생각하고,  
신입 대상 구인공고 데이터를 수집하여 "제안하는 연봉 및 근무조건이 전체적으로 지방보다 수도권이 우수한지" 를 분석하고자 한다.  

<br></br>

<b>해당 주제를 검증하기 위해 세 가지 가설을 설정하였다.  </b>

1. 구인 공고 데이터로 지역 예측이 가능할 것이다 (지도 학습).  

지역에 따라 일자리의 질적인 수준이 명확하게 나뉠수록, 구인 공고 데이터를 모아서 지역을 예측할 때 정확도가 높은 모델을 만들기 쉬울 것이다.

2. {수도권, 비수도권 광역시, 그 외 지방} 지역 중에서, <수도권> 지역이 '대기업 및 중견기업이 내건 구인공고 비중'이 가장 크고 <그 외 지방> 지역이 해당 공고 비중이 가정 작을 것이다.  

3. 구인 공고 데이터 피쳐를 바탕으로 클러스터링 시에  크게 그룹 {우수한 일자리 / 중간 / 상대적으로 별로인 일자리} 로  나눌  수 있을 것이며, 수도권에 가까울수록 우수한 일자리 그룹에 소속된 데이터포인트가 다수일 것이다. (비지도 학습)

만약 가설 3이 맞는다면, 수집한 데이터가 충분히 유의미한 정보를 포함하고 있다는 의미일 것이다.

### 분석 의의  

가설이 모두 참이라면, 단순히 지방지역의 청년층의 일자리 수 증가만 요구되는 것이 아니라  
청년층의 지방이탈을 방지하기 위해 양적인 측면에서의 개선도 필요함을 주장하는 근거 자료로 해당 분석 결과를 사용할 수 있다.
